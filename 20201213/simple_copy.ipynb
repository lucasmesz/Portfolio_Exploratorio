
Minimal Example
===============

Generating a square wordcloud from the US constitution using default arguments.


```python
import os

from os import path
from wordcloud import WordCloud

# get data directory (using getcwd() is needed to support running example in generated IPython notebook)
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

# Read the whole text.
#text = open(path.join(d, 'tese_doutor_marco_dias.txt')).read()
text=open('dissertacao_lucas.txt', encoding="utf8").read()

stop=open(r'stopwords.txt').read()
stopwords = set(stop)
stopwords.update(['contudo','entre','as','h','relao','ao','quando','dado','figura','tabela','dois','dos','seo','Captulo','sendo','j','t','','para', 'como', 'que', 'no', 'so', 'pode', 'ser','se','com','ou','no','uma','do','de','esse','um','Assim',
                 ' que',' de','que ','caso','tem', 'tambm','da','essa','em','seja','pois','na','mais','eq','isso', 'por', 'os'
                 , 'exemplo','forma','figura','mas','ex', 'problema', 'foi','apenas',
                 'aplicaes','outra','muito','dessa','onde','pelo','outro','mesmo'])
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords).generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
fig, ax=plt.subplots(figsize=(25,10))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")

# lower max_font_size
#wordcloud = WordCloud(max_font_size=40).generate(text)
#plt.figure()
#plt.imshow(wordcloud, interpolation="bilinear")
#plt.axis("off")
#plt.show()

# The pil way (if you don't have matplotlib)
# image = wordcloud.to_image()
# image.show()
```
